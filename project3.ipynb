# IQR

# Sepal_length iqr

```python
q1_sl=df.sepal_length.quantile(0.25)
q3_sl=df.sepal_length.quantile(0.75)
iqr_sl=q3_sl-q1_sl
iqr_sl
```

# Sepal_width iqr

```python
q1_sw=df.sepal_width.quantile(0.25)
q3_sw=df.sepal_width.quantile(0.75)
iqr_sw=q3_sw-q1_sw
iqr_sw
```

# petal_length iqr

```python
q1_pl=df.petal_length.quantile(0.25)
q3_pl=df.petal_length.quantile(0.75)
iqr_pl=q3_pl-q1_pl
iqr_pl
```

# petal_width iqr

```python
q1_pw=df.petal_width.quantile(0.25)
q3_pw=df.petal_width.quantile(0.75)
iqr_pw=q3_pw-q1_pw
iqr_pw
```

# Outlier detection

# outlier detection in sepal length

```python
UW=q3_sl +1.5 *iqr_sl
UW
LW=q1_sl-1.5*iqr_sl
LW
l=df['sepal_length']<LW
l_count=l.sum()
u=df['sepal_length']>UW
u_count=u.sum()
outlier=((df['sepal_length']<LW)|(df['sepal_length']>UW))
outlier.sum()
```

# outlier detection in sepal width

```python
UW=q3_sw +1.5 *iqr_sw
UW
LW=q1_sw-1.5*iqr_sw
LW
l=df['sepal_width']<LW
l_count=l.sum()
u=df['sepal_width']>UW
u_count=u.sum()
outlier=((df['sepal_width']<LW)|(df['sepal_width']>UW))
outlier.sum()
```

# outlier detection in petal length

```python
UW=q3_pl +1.5 *iqr_sw
UW
LW=q1_pl-1.5*iqr_sw
LW
l=df['petal_length']<LW
l_count=l.sum()
u=df['petal_length']>UW
u_count=u.sum()
outlier=((df['petal_length']<LW)|(df['petal_width']>UW))
outlier.sum()
```

# outlier detection in petal width

```python
UW=q3_pw +1.5 *iqr_sw
UW
LW=q1_pw-1.5*iqr_sw
LW
l=df['petal_width']<LW
l_count=l.sum()
u=df['petal_width']>UW
u_count=u.sum()
outlier=((df['petal_width']<LW)|(df['petal_width']>UW))
outlier.sum()
```
